```python
#creating a dataframe with required columns
df_sub=df_data[['Year','Count','Population','Crude Rate','Age_max']]
```

```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(df_sub.corr(), fignum=f.number) # type of plot
plt.xticks(range(df_sub.shape[1]), df_sub.columns, fontsize=14, rotation=45) # x axis labels
plt.yticks(range(df_sub.shape[1]), df_sub.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

## 3
Create a correlation matrix heatmap (colored).
Discuss what the coefficients in the plot mean in terms of the relationships between the variables.

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr_viz = sns.heatmap(df_sub.corr(), annot = True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
```

# insights gathered from heat map and correlation matrix
1. Year do not show any relation with any other variable as expected.
2. Count show weak positive relation with population
3. Count show strong positive relation with Crude rate . More incident  leads to higher incident rate
4. Population show weak negative relation with Crude rate. Population increases, Cancer incident rate decreases

## 6
Create a categorical plot and interpret the results.

```python
#different types of cancers agross both genders

sns.set(style="ticks")
categorical_viz_gen = sns.catplot(x="Count", y="Leading Cancer Sites", hue="Gender", data=df_data, height=4,aspect=2)
```

Biological impact gender:
Breast, Cervix Uteri,Corpus Uteri and Ovary cancers identified in Female
Prostarte Cancer is identified in only Male population

Thyroid and gallbladder cancers are found mostly in Female population.
Except Myeloma, Brain,Pancreas, all other types of Cancers are very high in Male population.

```python
# Create a categorical plot in seaborn using the leading cancer sites and Age

sns.set(style="ticks")
categorical_viz_age = sns.catplot(x="Count", y="Leading Cancer Sites", hue="Age_Flag", data=df_data,height=4, aspect=2)
```

Age seems to be have a huge impact on Cancer incidents.Spcifically middle-Elderly ages(50+ years)

```python
# Create a categorical plot in seaborn using the leading cancer sites and region

sns.set(style="ticks")
categorical_viz_region = sns.catplot(x="Count", y="Leading Cancer Sites", hue="Region", data=df_data,height=4, aspect=2)
```

West seems to have high counts in breast and Prostrate cancer cases. South has more cases of Lung and Bronchus Cancer.
Mid-west has less count of cases compared to other 3 regions.

```python
# Create a categorical plot in seaborn using the age groups, gender and leading cancer sites

sns.set(style="ticks")
categorical_viz_year= sns.catplot(x="Count", y="Leading Cancer Sites", hue="Year", data=df_data,height=4, aspect=2)
```

```python
# Create a categorical plot in seaborn using the age groups, gender and leading cancer sites

sns.set(style="ticks")
categorical_viz_year= sns.catplot(x="Age_max", y="Crude Rate", hue="Gender", data=df_data,height=4, aspect=2)
```

Shockingly, 2019 has more number of cancer incidents compared to 2018 also some extremely high number of cases are registered in 2019.
